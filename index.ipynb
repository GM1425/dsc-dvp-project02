```python
# Converting subcounty_crop_yield dataframe csv file to excel Karamoja.xlsx

subcounty_crop_yield.to_excel('Karamoja.xlsx', index = False)

```

```python
# Appending the second dataframe csv file to the created excel file Karamoja.xlsx

# specifying the Excel file and sheet name to append the dataframe to

excel_file = 'Karamoja.xlsx'
sheet_name = 'district_sheet'

# Writing the dataframe to the excel sheet

with pd.ExcelWriter(excel_file, mode = 'a') as writer:
    district_crop_yield.to_excel(writer, sheet_name = sheet_name, index = False)
    
```

```python
# Checking number of sheets in the Excel file to confirm the new sheet has been appended

# Creating a excel file panda object

xlsx_file = pd.ExcelFile ('Karamoja.xlsx')

# Extracting the list of sheet names

sheet_names = xlsx_file.sheet_names
sheet_names
```

```python
# getting the number of sheets in the excel file

len(sheet_names)

# this confirms we have the new sheet appended to our excel file.
```
